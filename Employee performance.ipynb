```python
# Function to detect, handle outliers, and plot boxplots
def handle_outliers_and_plot(data, columns):
    for column in columns:
        # Calculate Q1 (25th percentile) and Q3 (75th percentile)
        Q1 = data[column].quantile(0.25)
        Q3 = data[column].quantile(0.75)
        IQR = Q3 - Q1  # Interquartile Range

        # Define the outlier thresholds
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Plot boxplot before outlier treatment
        plt.figure(figsize=(10, 5))
        sns.boxplot(data[column])
        plt.title(f'Boxplot of {column} (Before Outlier Treatment)')
        plt.show()

        # Display the number of outliers
        outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
        print(f"Outliers in '{column}': {len(outliers)}")

        # Handle outliers by removing them
        data = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]

        # Plot boxplot after outlier treatment
        plt.figure(figsize=(10, 5))
        sns.boxplot(data[column])
        plt.title(f'Boxplot of {column} (After Outlier Treatment)')
        plt.show()

    return data

# Apply the function to handle outliers and plot boxplots
data_cleaned = handle_outliers_and_plot(data, numerical_columns)

# Display summary of the cleaned data
print("\nDataset after outlier treatment:")
print(data_cleaned.describe())
```
